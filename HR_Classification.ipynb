### ***HR Department Classification***



1.  # Importing Libraries And Dataset




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
employee_df=pd.read_csv('HR_DataSet.csv')
employee_df.head()
```

```python
employee_df.describe()
```

# 2.  **Data PreProcessing**
## Label Encoding For Output



```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
employee_df['Attrition'] = le.fit_transform(employee_df['Attrition'])
employee_df['OverTime'] = le.fit_transform(employee_df['OverTime'])
employee_df
```

## Ensuring There Is No Null Values

```python
sns.heatmap(employee_df.isnull(),yticklabels=False,cbar=False,cmap='Blues')
```

## Analyzing Data And Figure Out Which Data To be Dropped

```python
employee_df.hist(bins=30,figsize=(25,25), color='r')
```

## Dropping Columns

```python
employee_df.drop(['EmployeeCount','StandardHours', 'EmployeeNumber','Over18'],axis=1,inplace=True)
employee_df
```

## Comparing Number of Employees Who Stayed Or Left

```python
left_df=employee_df[employee_df['Attrition']==1]
stayed_df=employee_df[employee_df['Attrition']==0]
print("number of employee who left =", len(left_df))
print("number of employee who stayed =", len(stayed_df))
print("percentage of employee left=",len(left_df)/len(employee_df)*100,"%")
print("percentage of employee stayed=",len(stayed_df)/len(employee_df)*100,"%")
```

## Correlation Between Variables

```python

correlations = employee_df.select_dtypes(include=['number']).corr()
f, ax = plt.subplots(figsize = (22, 22))
sns.heatmap(correlations, annot = True)
```

## Dealing With Categorical Variables

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder()

columns_to_encode = ['JobRole', 'Department', 'EducationField', 'Gender', 'BusinessTravel', 'MaritalStatus']
encoded_data = enc.fit_transform(employee_df[columns_to_encode]).toarray()

feature_names = enc.get_feature_names_out(columns_to_encode)

encoded_df = pd.DataFrame(encoded_data, columns=feature_names)


employee_df = pd.concat([employee_df, encoded_df], axis=1)

employee_df.drop(columns_to_encode, axis=1, inplace=True)


employee_df.head()
```

## Splitting the data into input and output and starting model training

```python
X=employee_df.drop('Attrition',axis=1)
y=employee_df['Attrition']
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

## Feature Scaling Using Normalization

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```



# 3.   Building Models



## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print("Accuracy {} %".format( 100 * accuracy_score(y_pred, y_test)))
```

```python
# Testing Set Performance
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```

## Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```

## ANN Classification

```python
from keras.models import Sequential
from keras.layers import Dropout,Dense
from tensorflow.keras.regularizers import l2
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)
```

```python
model=Sequential([
    Dense(units=200,input_dim=X_train.shape[1],kernel_regularizer=l2(0.01),activation='relu'),
    Dropout(0.50),
    Dense(units=200,activation='relu',kernel_regularizer=l2(0.01)),
    Dropout(0.30),
    Dense(units=100,activation='relu',kernel_regularizer=l2(0.01)),
    Dropout(0.25),
    Dense(units=50,activation='relu'),
    Dropout(0.20),
    Dense(units=1,activation='sigmoid'),
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train,y_train,batch_size=32,epochs=100,validation_split=0.2, callbacks=[early_stopping])
```

```python
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True)
```
